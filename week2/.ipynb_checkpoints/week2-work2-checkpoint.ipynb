## 【課題1】1次元配列を作ろう

問題に、データは[0, 0.01, 0.02・・・0.98, 0.99, 1]とあるため、データは101個となる

```python
import numpy as np

x = np.arange(0, 1.01, 0.01)
```

```python
x
```

## 【課題2】2次元配列を作ろう

```python
list = np.zeros((0,2))

for i in x:
    y = i/2 + 1
    list = np.append(list,[[i, y]], axis=0)
```

```python
#　一次関数(y = x/2 + 1)のxとyの値
list
```

## 【課題3】勾配を求めよう

勾配は (yの変化量) / (xの変化量)
{y(n+1) - y(n)} / {x(n+1) - x(n)}
で求めることにする。
また、データが101個となるため100個の勾配が出る。

```python
# 変化量を求める。
for i in range(len(list)-1):
    linear_func_gradient = (list[i+1][1]-list[i][1]) / (list[i+1][0]-list[i][0])
    print(linear_func_gradient)
```

## 【課題5】最小値の求め方を説明しよう

二次関数の最小値を求める方法は、  
隣り合うxを数個選択する。（例えば、x(n-1)、x(n)、x(n+1)の3点）  
x(n-1)とx(n)の勾配およびx(n)、x(n+1)の勾配を求める。  
２つの勾配を比べ、小さい方を選択する。  
小さい勾配がx(n-1)とx(n)であれば、次にx(n-2)x(n-1)とx(n)の3点で同様に勾配を求める。  
上記を繰り返し勾配が０になる箇所を探す。  

## 【課題6】最小値を求めよう

y = x**2 + 5x　を微分すると

y' = 2x + 5になる。  
上記からy' = 0となるxは１つがわかる。  
xを求めると-5/2となり、その時のyは-25/4  
よって最小値は、-25/4となる。
