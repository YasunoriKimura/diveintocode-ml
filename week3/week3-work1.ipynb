## 【課題1】データの取得

```python
from sklearn.datasets import load_iris
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = load_iris()
```

```python
X = DataFrame(data=data.data, columns=pd.Index(["sepal_length", "sepal_width", "petal_length", "petal_width"]))
```

```python
y = DataFrame(data=data.target, columns=pd.Index(["Species"]))
```

```python
y
```

## 【課題2】データの結合

```python
df = X.join(y)
```

```python
df
```

## 【課題4】irisデータセット自体の調査

irisデータセットは、英国の統計学者・植物学者であったロナルド・フィッシャーが1936年に論文に発表した多変量データ。  
データセットは、アイリスの花の3種類(「Iris setosa」と「 Iris virginica」 と 「Iris versicolor」) を  
各々50サンプルずつ集めた計150のデータからなり、変数は、がく片の長さと幅、花びら長さと幅をcmで計測した4つの変量からなっている。 

## 【課題5】必要なデータを抽出

  .loc : 指定方法がラベル名  
  .iloc : 指定方法がインデックス番号
### ・50番目から99番目までのデータを抽出する  

```python
df["sepal_width"][50:100]
```

```python
df.iloc[50:100, [1]]
```

---
### ・petal_lengthの列の50番目から99番目までのデータを抽出する

```python
df.loc[50:100, ["petal_length"]]
```

---
### ・petal_widthの値が0.2のデータを抽出する

```python
df[df['petal_width'].isin([0.2])]
```

## 【課題7】特徴量間の関係の確認

### ・縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
df.plot(kind='scatter', x=df.columns[0], y=df.columns[2], c="Species")
```

***
### ・全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

```python
sns.pairplot(df, hue='Species')
```

---
### ・４つの特徴量に関する相関係数行列を作る

```python
corr_mat = df.corr(method='pearson')
```

```python
corr_mat
```

---
### ・相関係数行列のヒートマップを作る

```python
sns.heatmap(corr_mat,
            annot=True,
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
plt.show()
```

## 【課題8】簡単な考察
用語
ガクの長さ(sepal_length)、幅(sepal_width)、花弁の長さ(petal_length)、幅(petal_width)

setosaは、３種の中で花弁の長さが一番短く(0〜2cmほど)、幅が狭い(0〜1cm)。額の長さは0〜6cmほど。  
versicolorは、花弁の長さが3〜6cmほど、花弁幅が1〜2cmでsetosaとvirginicaの中間くらい。  
virginicaは、花弁の長さと花弁幅が３種の中で一番長い。
