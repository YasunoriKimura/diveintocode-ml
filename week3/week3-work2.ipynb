## 【課題5】用語の調査

・尖度  
　分布をグラフで表した場合のコブの尖り具合を示している。  
　尖度が大きくなるにつれて、より尖った分布になる。  
・歪度  
　分布の歪み具合を示している。グラフで表すと歪度>0の時は左側にコブがある。歪度=0の時は左右対称。歪度<0は右側にコブが分布になる。

## 【課題6】分布の確認

### ・seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

```python
sns.distplot(df_train.SalePrice)
```

```python
#　目的変数分布での尖度
df_train.SalePrice.kurtosis()
```

```python
#　目的変数分布での歪度
df_train.SalePrice.skew()
```

---
### ・目的変数に対して対数変換を行う。  

```python
#　　対数変換
np.log(df_train.SalePrice)
```

---
### ・対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
sns.distplot(np.log(df_train.SalePrice))
```

```python
# 目的変数に対して対数変換後の尖度
np.log(df_train.SalePrice).kurtosis()
```

```python
#　# 目的変数に対して対数変換後の歪度
np.log(df_train.SalePrice).skew()
```

対数変換前後では、尖度および歪度が0に近づいているため、正規分布近い分布となっている。

## 【課題7】相関係数の確認

#### ・相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

```python
# 特徴量に関する相関係数行列を作る
corr_mat = df_train.corr(method='pearson')
```

```python
corr_mat
```

```python
sns.heatmap(corr_mat, square=True)
```

---
#### ・目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
# 目的変数との相関の高い特徴量10個抽出
corr_mat.nlargest(10, 'SalePrice')['SalePrice']
```

```python
"""
目的変数との相関の高い特徴量10個の
相関係数行列ヒートマップを作成する。
"""

#　 抽出した10個の特徴量ラベル
t10_labels = corr_mat.nlargest(10, 'SalePrice')['SalePrice'].index

#  相関係数行列
corr_mat_t10_labels = df_train[t10_labels].corr(method='pearson')

#　ヒートマップ作成
sns.heatmap(corr_mat_t10_labels, annot=True, fmt='.2f', annot_kws={"size": 9}, square=True)
```

---
