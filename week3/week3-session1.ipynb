### 欠損が○%以上の列は削除、それ以外の欠損値を「0」「平均値」「中央値」「最頻出値」で埋めた場合の関数作成  

まずはそれぞれ関数を作成してみる。

input 欠損値数の許容パーセンテージ

def　make_heatmap〜(許容値)：
    許容値以上の列を削除
    残りを埋める。
    相関行列を作る。
    top10を出し再度相関行列を作成
    ヒートマップ作成
    
output　グラフ

```python
def make_heatmap_0(permission_num_of_missing_value):
    """
    input%以上が欠損しているカテゴリーは削除して、それ以外の欠損値を0で埋める関数
    """
    
    # input数以上が欠損しているカテゴリーは削除
    df_train_less = dummy_df_train.dropna(thresh=round(307511*(1-permission_num_of_missing_value/100)), axis=1)

    # 残りの欠損値を「平均値」で埋める
    df_train_less_0 = df_train_less.fillna(df_train_less.fillna(0))
    
    # 特徴量に関する相関係数行列を作る
    corr_mat_less_0 = df_train_less_0.corr(method='pearson')

    # 目的変数との相関の高い特徴量10個抽出し、カテゴリー名を格納
    less_t10_0 = corr_mat_less_0.nlargest(10, 'TARGET')['TARGET'].index
    
    #  相関係数行列
    corr_mat_less_t10_0 = df_train_less_0[less_t10_0].corr(method='pearson')
    
    # ヒートマップ作成
    heatmap_less_t10_0 = sns.heatmap(corr_mat_less_t10_0, annot=True, fmt='.2f', annot_kws={"size": 9}, square=True)
    
    return heatmap_less_t10_0


def make_heatmap_mean(permission_num_of_missing_value):
    """
    input%以上が欠損しているカテゴリーは削除して、それ以外の欠損値を平均で埋める関数
    """

    # input数以上が欠損しているカテゴリーは削除
    df_train_less = dummy_df_train.dropna(thresh=round(307511*(1-permission_num_of_missing_value/100)), axis=1)

    # 残りの欠損値を「平均値」で埋める
    df_train_less_mean = df_train_less.fillna(df_train_less.mean())
    
    # 特徴量に関する相関係数行列を作る
    corr_mat_less_mean = df_train_less_mean.corr(method='pearson')

    # 目的変数との相関の高い特徴量10個抽出し、カテゴリー名を格納
    less_t10_mean = corr_mat_less_mean.nlargest(10, 'TARGET')['TARGET'].index
    
    #  相関係数行列
    corr_mat_less_t10_mean = df_train_less_mean[less_t10_mean].corr(method='pearson')
    
    # ヒートマップ作成
    heatmap_less_t10_mean = sns.heatmap(corr_mat_less_t10_mean, annot=True, fmt='.2f', annot_kws={"size": 9}, square=True)
    
    return heatmap_less_t10_mean


def make_heatmap_median(permission_num_of_missing_value):
    """
    input%以上が欠損しているカテゴリーは削除して、それ以外の欠損値を中央値で埋める関数
    """

    # input数以上が欠損しているカテゴリーは削除
    df_train_less = dummy_df_train.dropna(thresh=round(307511*(1-permission_num_of_missing_value/100)), axis=1)

    # 残りの欠損値を「中央値」で埋める
    df_train_less_median = df_train_less.fillna(df_train_less.median())
    
    # 特徴量に関する相関係数行列を作る
    corr_mat_less_median = df_train_less_median.corr(method='pearson')

    # 目的変数との相関の高い特徴量10個抽出し、カテゴリー名を格納
    less_t10_median = corr_mat_less_median.nlargest(10, 'TARGET')['TARGET'].index
    
    #  相関係数行列
    corr_mat_less_t10_median = df_train_less_median[less_t10_median].corr(method='pearson')
    
    # ヒートマップ作成
    heatmap_less_t10_median = sns.heatmap(corr_mat_less_t10_median, annot=True, fmt='.2f', annot_kws={"size": 9}, square=True)
    
    return heatmap_less_t10_median


def make_heatmap_most(permission_num_of_missing_value):
    """
    input%以上が欠損しているカテゴリーは削除して、それ以外の欠損値を最頻出値で埋める関数
    """

    # input数以上が欠損しているカテゴリーは削除
    df_train_less = dummy_df_train.dropna(thresh=round(307511*(1-permission_num_of_missing_value/100)), axis=1)

    # 残りの欠損値を「再頻出値」で埋める
    df_train_less_most = df_train_less.fillna(df_train_less.mode().iloc[0])
    
    # 特徴量に関する相関係数行列を作る
    corr_mat_less_most = df_train_less_most.corr(method='pearson')

    # 目的変数との相関の高い特徴量10個抽出し、カテゴリー名を格納
    less_t10_most = corr_mat_less_most.nlargest(10, 'TARGET')['TARGET'].index
    
    #  相関係数行列
    corr_mat_less_t10_most = df_train_less_most[less_t10_most].corr(method='pearson')
    
    # ヒートマップ作成
    heatmap_less_t10_most = sns.heatmap(corr_mat_less_t10_most, annot=True, fmt='.2f', annot_kws={"size": 9}, square=True)
    
    return heatmap_less_t10_most
```

#### ランダムフォレストを使い特徴量の重要度を算出

```python
#  欠損値が10%以上の列は削除
df_train_less10 = dummy_df_train.dropna(thresh=276760, axis=1)

# 残りの欠損値を「平均値」で埋める
df_train_less10_mean = df_train_less10.fillna(df_train_less10.mean())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

(X_train, X_test, 
     y_train, y_test) = train_test_split(df_train_less10_mean.drop("TARGET", axis=1), df_train_less10_mean.TARGET, test_size=0.1)

clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)
y_pred = clf_rf.predict(X_test)

accu = accuracy_score(y_test, y_pred)
print('accuracy = {:>.4f}'.format(accu))

# Feature Importance
fti = clf_rf.feature_importances_

feature_importance = {}
for i, feat in enumerate(X_train.columns):
    feature_importance[feat] = round(fti[i], 3)
```

```python
df_feature_importance = pd.DataFrame.from_dict(feature_importance, orient='index', columns=['value(%)'])
```

```python
df_feature_importance.plot(
    kind='barh', figsize=(6,45), title='feature importance from RTF(missing values less than 10)')
```

欠損値10%以下を削除し、残りを平均値で埋めたもので重要度を算出したが、重要度は最高値でも0.07程度であった。

#### 欠損値が50%以上を削除で実施

```python
#  欠損値が50%以上の列は削除
df_train_less50 = dummy_df_train.dropna(thresh=round(307511/2), axis=1)

# 残りの欠損値を「平均値」で埋める
df_train_less50_mean = df_train_less50.fillna(df_train_less50.mean())

(X_train, X_test, 
     y_train, y_test) = train_test_split(df_train_less50_mean.drop("TARGET", axis=1), df_train_less50_mean.TARGET, test_size=0.1)

clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)
y_pred = clf_rf.predict(X_test)

accu = accuracy_score(y_test, y_pred)
print('accuracy = {:>.4f}'.format(accu))

# Feature Importance
fti = clf_rf.feature_importances_

feature_importance = {}
for i, feat in enumerate(X_train.columns):
    feature_importance[feat] = round(fti[i], 3)
    
df_feature_importance = pd.DataFrame.from_dict(feature_importance, orient='index', columns=['value(%)'])

df_feature_importance.plot(
    kind='barh', figsize=(6,45), title='feature importance from RTF(missing values less than 10)')
```

まだ1%超えるものはなかった。

#### lightGBMを使用し、重要度を算出

```python
import lightgbm as lgb
```

```python
# TARGET列をyへ移行
X = dummy_df_train.drop("TARGET",axis=1)
y = dummy_df_train.TARGET

# トレーニングデータとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y)

# データセットを生成する
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)

# LightGBM のハイパーパラメータ
lgbm_params = {
    # ２クラス問題
    'objective': 'binary',
    'metric': 'auc',
}

# 上記のパラメータでモデルを学習する
model = lgb.train(lgbm_params, lgb_train, valid_sets=lgb_eval)
```

```python
#　重要度グラフ
lgb.plot_importance(model, figsize=(10,40))
```

上記より、EXT_SOURCE_1　EXT_SOURCE_2　EXT_SOURCE_3　DAYS_BIRTH...が高いことがわかった。  
上位３つは外部データからの正規化スコアであった。  
ローンの金額や年金ローンの重要度が高いことは理解できるが、誕生日も重要度が高いことには驚きであった。

---

```python
# 重要度の数値算出
feat_importance = model.feature_importance()

# 重要度一覧をデータフレーム型に変換し上位20項目を算出
df_feat_importance = pd.DataFrame({
    "columns": X_train.columns,
    "feat_importance": feat_importance,
}).sort_values('feat_importance', ascending=False).head(20)
```
