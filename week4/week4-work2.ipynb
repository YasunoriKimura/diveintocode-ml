# Week4授業前課題2 住宅価格を回帰分析しよう

## 【課題1】練習のために特徴量の選択

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_train = pd.read_csv('train.csv')
```

```python
# 説明変数と目的変数に分割
X = df_train[['GrLivArea', 'YearBuilt']].values
y = df_train[['SalePrice']].values

# トレーニングデータとテストデータに分割
(X_train, X_test, 
     y_train, y_test) = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

## 【課題2】回帰問題のパイプライン

```python
#　学習
clf_reg = LinearRegression().fit(X_train, y_train)

#推定
y_predict_reg = clf_reg.predict(X_test)
```

```python
#評価
mean_squared_error(y_test, y_predict_reg)
```

```python
#予測値
y_predict_reg.shape
```

```python
#グラフ描画
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.set_title('Relationship between GrLivArea and SalesPrice')
ax1.set_xlabel('GrLivArea')
ax1.set_ylabel('SalesPrice')
ax1.scatter(x=X_test[:, 0], y=y_predict_reg)

ax2.set_title('Relationship between YearBuilt and SalesPrice')
ax2.set_xlabel('YearBuilt')
ax2.set_ylabel('SalesPrice')
ax2.scatter(x=X_test[:, 1], y=y_predict_reg)

plt.show()
```

## 【課題3】手法の比較

### ・SVMによる評価

```python
from sklearn.svm import SVR

#学習
clf_svr = SVR().fit(X_train, y_train)

#推定
y_predict_svr = clf_svr.predict(X_test)

#評価
mean_squared_error(y_test, y_predict_svr)
```

```python
y_predict_svr
```

### ・決定木による評価

```python
from sklearn.tree import DecisionTreeRegressor

#学習
clf_dtr = DecisionTreeRegressor().fit(X_train, y_train)

#推定
y_predict_dtr = clf_dtr.predict(X_test)

#評価
mean_squared_error(y_test, y_predict_dtr)
```

### ・ランダムフォレストによる評価

```python
from sklearn.ensemble import RandomForestRegressor

#学習
clf_rfr = RandomForestRegressor().fit(X_train, y_train)

#推定
y_predict_rfr = clf_svr.predict(X_test)

#評価
mean_squared_error(y_test, y_predict_rfr)
```

```python
y_predict_rfr
```

## それぞれの手法の可視化

```python
def make_scatters(y_predict):
    #グラフ描画
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.set_title('Relationship between GrLivArea and SalesPrice')
    ax1.set_xlabel('GrLivArea')
    ax1.set_ylabel('SalesPrice')
    ax1.scatter(x=X_test[:, 0], y=y_predict)

    ax2.set_title('Relationship between YearBuilt and SalesPrice')
    ax2.set_xlabel('YearBuilt')
    ax2.set_ylabel('SalesPrice')
    ax2.scatter(x=X_test[:, 1], y=y_predict)
    
    plt.show()
```

#### SVM

```python
make_scatters(y_predict_svr)
```

#### 決定木

```python
make_scatters(y_predict_dtr)
```

#### ランダムフォレスト

```python
make_scatters(y_predict_rfr)
```

### 考察

グラフからは線形回帰と決定木はうまく分布されているが、SVMとランダムフォレストは一定になっていると読み取れる。  
SVMとランダムフォレストの予測値を確認してもほぼ一定の予測価格となっている為、グラフと一致する。  
また、SVMとランダムフォレストの予測価格と実価格に大きな誤差があるため、平均二乗誤差も高い値となっている。これも一致する。  
分類問題と回帰問題の両方で使えない手法とは、SVMとランダムフォレストではないかと考えられる。  

線形回帰と決定木のグラフから、当たり前ではあるが広いほど高値がつけられている。
新しいより広いほうが高値をつけられる傾向もある。
